```python
# project 2: calculator

def add(x,y):
    return x+y

def sub(x,y):
    return x-y

def multi(x,y):
    return x*y

def div(x,y):
    return x/y
```

```python
print("Enter A for addition: ")
print("Enter S for Subtraction: ")
print("Enter M for Multiplication: ")
print("Enter D for Division: ")


choice = input('Enter choice (Add, Sub, Mul, Div)')



if choice.upper() in ('A', 'ADD', 'ADDITION', 'S', 'SUBTRACT', 'M', 'MULTIPLY', 'MULTIPLICATION', 'DIVISION', 'D', 'DIVIDE'):

    num1 = float(input('enter 1st number: '))

    num2 = float(input('enter 2nd number: '))

    if choice.upper() in ('A', 'ADD', 'ADDITION'):
        print('Result: 'num1, '+', num2, ' = ', add(num1,num2))

    elif choice.upper() in ('S', 'SUBTRACT'):
        print('Result: 'num1, '-', num2, ' = ', sub(num1,num2))

    elif choice.upper() in ('M', 'MULTIPLY', 'MULTIPLICATION'):
        print('Result: 'num1, '*', num2, ' = ', multi(num1,num2))

    elif choice.upper() in ('D', 'DIVIDE', 'DIVISION'):
        print('Result: 'num1, '/', num2, ' = ', div(num1,num2))

else:
    print("enter a right choice")
    
```
